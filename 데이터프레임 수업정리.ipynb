```python
df_samsung=pd.read_excel('D:\정유재\금융데이터 분석을 위한 판다스 활용법\자료\Data\samsung_kospi_excel.xlsx',sheet_name='samsung',index_col=0)
df_kospi=pd.read_excel('D:\정유재\금융데이터 분석을 위한 판다스 활용법\자료\Data\samsung_kospi_excel.xlsx',sheet_name='kospi',index_col=0)
```

```python
df=pd.DataFrame() #빈 df 만들기

df['Kospi_Return'] = df_kospi['Change'] # 빈 df에 kospi return이라는 칼럼을 만들고, 그 칼럼에 정보는 kospi의 수익률 데이터로 채운다
                                        # df[칼럼]=df[데이터]
df['Samsung_Return']=df_samsung['Change']
```

```python
#조건: kospi 2%상승 and 삼전 2%상승한 날은?

df['Kospi_Return']>=0.02
df['Samsung_Return']>0.02

(df['Kospi_Return']>=0.02) & (df['Samsung_Return']>0.02) # and-> 논리 연산자. &-> 비교 연산자
```

```python
df['Condition']=np.where(((df['Kospi_Return']>=0.02) & (df['Samsung_Return']>0.02)), df['Samsung_Return'], np.nan) #조건문 작성해주고, 일치, 불일치

(df['Condition'].dropna()).index # 저 날들의 인덱스값만 추출하고 싶을 때

df_result=df.dropna()

df_result.sort_index(ascending=False) #index 기준 sorting
df_result.sort_values(by='Condition',ascending=False) #value 기준 sorting/ value는 by 조건을 만들어줘야함
```

```python
import pandas as pd
#데이터파일 정제하기
df=pd.read_excel('D:\정유재\금융데이터 분석을 위한 판다스 활용법\자료\Data\groupby_example.xlsx',header=6) #더러운 파일이기에 정제를 해줘야합 header를 찾고 ㄴ
df=df.dropna(axis=1) #nan값 없애주고
df.columns=['기업','Sector','PER'] # 칼럼값 다시 주고
df.index=df['기업'] # 인덱스이름을 기업을 바꿔주고
df=df.drop('기업',axis=1) #칼럼에 있는 기업 줄 쫙 버려주고. axis=1 세로 줄 떨구기 axis=0 가로 줄 없애기

df
```

```python
df.groupby(['Sector']).mean() #섹터별로 묶어주고 , mean count 할 수 있음

undervalued_mean=df.groupby(['Sector']).mean().loc['반도체'].values[0] # 그룹핑 해놓은 sector의 평균값에서 '반도체' 라는 인덱스를 불러오는 것이기에 loc[인덱스 이름] 으로 그 값을 도출했다.
                                                                        # 도출한 값의 value, 즉 값을 불러오는 것이기에 values를 해준 것이고, array 값으로 나왔으니까 [0]을 해주어서 값만 도출.

undervalued_mean
```

```python
df_mi=df[df['Sector']=='반도체'] #sector가 반도체인것 것들을 df 형태로 가져옴
df_mi['Condition']=np.where(df_mi['PER']<=undervalued_mean,1,np.nan) # 새로운 컨디션이라는 칼럼을 만들고 거기에 새로운 데이터를 추가
df_mi.dropna() #Nan 값 버려버리기 목표로 하는 기업들 도출 가능
```
